```python
import pandas as pd
import numpy as np
```

```python
books=pd.read_csv('books.csv',encoding ='latin1')
books.head()
books.shape
```

```python
books.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
books=books.rename({'User.ID':'UserId'},axis=1)
```

```python
books=books.rename({'Book.Title':'Title','Book.Rating':'Rating'},axis=1)
```

```python
books
```

```python
books=books.sort_values(by=['UserId'])
```

```python
#no of unique users in the dataset
len(books.UserId.unique())
```

```python
len(books.Title.unique())
```

```python
book_user=books.pivot_table(index='UserId',columns='Title',values='Rating').reset_index(drop=True)
```

```python
book_user
```

```python
book_user=book_user.fillna(0)
```

```python
book_user.index=books.UserId.unique()
```

```python
book_user
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( book_user.values,metric='cosine')
```

```python
user_sim
```

```python
#store the results in a dataframe
book_df=pd.DataFrame(user_sim)
book_df
```

```python
#set the index and column names to user ids
book_df.index=books.UserId.unique()
book_df.columns=books.UserId.unique()
```

```python
book_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(book_df,0)
```

```python
#Most Similar Users
book_df.idxmax(axis=1)[0:1348]
```
